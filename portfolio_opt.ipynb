```python
import yfinance as yf
import pandas as pd

# Define stock symbols
stock_symbols = ['HDFCBANK.NS', 'TCS.NS', 'ICICIBANK.NS', 'INFY.NS',
                 'BAJFINANCE.NS', 'MARUTI.NS', 'HINDUNILVR.NS', 'ASIANPAINT.NS',
                 'DRREDDY.NS', 'EICHERMOT.NS', 'CHOLAFIN.NS',
                  'ASHOKLEY.NS', 'GESHIP.NS',
                 'BRITANNIA.NS']

# Download data from Yahoo Finance
data = yf.download(stock_symbols, start='2005-01-01', end='2024-01-01')

# Extracting closing prices
closing_prices = data['Close']

# Reshape the DataFrame to have date as index and stock symbols as columns
closing_prices = closing_prices.stack().unstack(level=1)

closing_prices.head()
```

```python
returns=closing_prices.pct_change()
```

```python
returns=returns.iloc[1:,:]
```

```python
returns.describe()
```

```python
covariance_matrix=returns.cov()
covariance_matrix
```

```python
def weights(c):
  one=np.ones(len(c))
  inv_cov=inv(c)
  p=np.dot(inv_cov,one.T)
  return (np.dot(one,inv_cov))/(np.dot(one,p))
```

```python
minimum_var_weights=weights(covariance_matrix)
minimum_var_weights=np.round(minimum_var_weights*100,3)
print("minimum weight matrix of 14 stocks : ",minimum_var_weights)
```

```python
minimum_var_weights.sum()
```

```python
weight_df = pd.DataFrame({'Stock(Ticker)': stock_symbols, 'Weight (%)': minimum_var_weights})
weight_df
```

```python
sns.barplot(data=weight_df,x=stock_symbols,y=minimum_var_weights)
```
